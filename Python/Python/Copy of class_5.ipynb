```python
R=int(input("Enter the number: "))

if(R>=90):
    print("Grade: A")
elif(R>=80 & R<90):
    print("Grade: B")
elif(R>=70 & R<80):
    print("Grade: C")
elif(R >=60 & R<70):
    print("Grade: D")
else:
    print("Grade: F")
```

```python
R = int(input("Enter the number: "))

match R:
    case R if R >= 90:
        print("Grade: A")
    case R if 80 <= R < 90:
        print("Grade: B")
    case R if 70 <= R < 80:
        print("Grade: C")
    case R if 60 <= R < 70:
        print("Grade: D")
    case _:
        print("Grade: F")
```

```python
 number = int(input("Enter a number: "))
 is_prime = True
 if number > 1:
  for i in range(2, int(number ** 0.5) + 1):
     if number % i == 0:
       is_prime = False
       break
     if is_prime:
        print(f"{number} is a prime number.")
     else:
        print(f"{number} is not a prime number.")
 else:
  print(f"{number} is not a prime number.")
```

```python
 n = int(input("Enter the number of terms in Fibonacci sequence: "))
 a, b = 0, 1
 print("Fibonacci sequence:")
 for _ in range(n):
   print(a, end=' ')
   a, b = b, a + b
 print()
```
